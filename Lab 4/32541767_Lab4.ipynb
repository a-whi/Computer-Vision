## Linear Regression

_Linear regression_, as the name implies, finds the _best fit line_ (or plane/hyperplane in high-dimensional spaces) between the _independent_ and _dependent_ variable (i.e., input data and target output). Recall that we approached the univariate linear regression in the class first and built our way up to multivariate input/output forms. Univariate linear regression describes the situation where only one independent variable is presented (i.e., $x \in \mathbb{R}$) and the model has to find the linear relationship of it with the dependent variable (i.e., $y$). In the multivariate form, we have more than one independent variable (i.e., $\boldsymbol{x} \in \mathbb{R}^n$) and the model may have multiple outputs to predict (i.e., $\boldsymbol{y} \in \mathbb{R}^p$). <br>

Here $m$ is the number of data samples we have, $n$ is the dimensionality of input $\boldsymbol{x}$ and $p$ is the dimensionality of target $\boldsymbol{y}$. <br>You may **not** use  pre-existing linear regression functions (e.g., the one from scikit-learn). <br>

#### 1. Regression function

```python
def linearRegression_cof(X, Y):
    if X.ndim == 1:
        X = X.reshape(-1, 1)  # make sure it's (m, n)
    
    # Add bias term: augment with column of ones
    ones = np.ones((X.shape[0], 1))
    X_aug = np.hstack((ones, X))  # shape becomes (m, n+1)

    # Compute coefficients using pseudo-inverse
    W = np.linalg.pinv(X_aug.T @ X_aug) @ X_aug.T @ Y

    return W
```

#### 4. & 5. Predicting the Life Expectancy

Write a function **predict()**, which takes a data point $\boldsymbol{x}_q$ and coefficients $\boldsymbol{w}$ and return the prediction $\hat{y} = \boldsymbol{w}^\top \boldsymbol{x}_q$.

```python
def predict(X, w):
    # Ensure X is 2D
    if X.ndim == 1:
        X = X.reshape(1, -1)  # Convert a single sample to shape (1, n)
    
    # Augment X with a column of ones (for bias)
    ones = np.ones((X.shape[0], 1))
    X_aug = np.hstack((ones, X))  # Shape: (m, n+1)
    
    # Compute prediction
    Y_hat = X_aug @ w  # Shape: (m, p)

    return Y_hat
```

Now, we will use this **predict()** function to plot Y = 'Life expectancy' against the predictor X = 'Alcohol' and 'BMI'. You are asked to display both the original data as a scatter plot and your predicted best fit line.

```python
# For Alcohol
w_alc = linearRegression_cof(X_alc, Y)
Y_pred_alc = predict(X_alc, w_alc)

# Sort for plotting a clean line
sorted_idx = X_alc[:, 0].argsort()
X_sorted = X_alc[sorted_idx]
Y_sorted = Y_pred_alc[sorted_idx]

plt.figure(figsize=(8, 5))
plt.scatter(X_alc, Y, label='Original data', color='blue')
plt.plot(X_sorted, Y_sorted, label='Best fit line (Alcohol)', color='red')
plt.xlabel('Alcohol')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy vs Alcohol')
plt.legend()
plt.grid(True)
plt.show()

# Fo BMI
w_bmi = linearRegression_cof(X_bmi, Y)
Y_pred_bmi = predict(X_bmi, w_bmi)

sorted_idx = X_bmi[:, 0].argsort()
X_sorted = X_bmi[sorted_idx]
Y_sorted = Y_pred_bmi[sorted_idx]

plt.figure(figsize=(8, 5))
plt.scatter(X_bmi, Y, label='Original data', color='green')
plt.plot(X_sorted, Y_sorted, label='Best fit line (BMI)', color='orange')
plt.xlabel('BMI')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy vs BMI')
plt.legend()
plt.grid(True)
plt.show()
```

#### 2. Computing the Error to evaluate the performance of the linear regression models

```python
#Compute the sum squared error
def compute_sse(y_estimate, y): 
    # Ensure inputs are numpy arrays
    y_estimate = np.array(y_estimate)
    y = np.array(y)
    
    # Compute squared differences and sum them
    sse = np.sum((y - y_estimate) ** 2)

    return sse
```

## Task 4: Peak Signal-to-Noise Ratio PSNR

The _Peak Signal to Noise Ration_ (PSNR) is defined as the ratio of the maximum possible power of a signal to the power of corrupting noise that affects representation fidelity. PSNR is most easily defined via the mean squared error (MSE). Given a noise-free $m\times n$ grayscale image $I$ and its noisy approximation $K$, the $\mathrm{MSE}$ and $\mathrm{PSNR}$ are defined as

\begin{align}
    \mathrm{MSE} &= \frac{1}{mn} \sum^{m-1}_{i=0} \sum^{n-1}_{j=0} [I(i,j)-K(i,j)]^{2} \\
    \mathrm{PSNR} &= 20 \cdot log_{10}(\mathrm{MAX}_I))-10 \cdot log_{10}(\mathrm{MSE}).
\end{align}

Note that $\mathrm{MAX}_I$ denotes the maximum possible value a pixel can have.

See this [Reference](https://en.wikipedia.org/wiki/Peak_signal-to-noise_ratio) for more details.

#### 1. PSNR Implementation

```python
## Implement PSNR to measure quantitative difference
def PSNR(img1, img2):
    # Ensure the images are in [0, 255] and type is integer
    img1 = np.array(img1, dtype=np.uint8)
    img2 = np.array(img2, dtype=np.uint8)

    # Convert to float for computation to avoid overflow/underflow
    img1 = img1.astype(np.float64)
    img2 = img2.astype(np.float64)
    
    # Compute Mean Squared Error
    mse = np.mean((img1 - img2) ** 2)
    
    if mse == 0:
        return float('inf')  # Perfect match

    # Maximum pixel value
    max_pixel = 255.0
    
    # Compute PSNR
    psnr = 20 * np.log10(max_pixel / np.sqrt(mse))
    
    return psnr
```

**Main idea** (concise): Parse the image and check for pixels whose value is exactly 0 (i.e. black corrupted pixel);  For each of these pixels, take the area right above in a 5x11 pixel region from the corresponding uncorrupted training image and add it to the samples list (like we did before with the entire images) -- the Y value is the corresponding intensity of the actual pixel to be reconstructed (also to be extracted from the clean image).

#### 1. Patch extraction
Create a function to extract a patch of size 'patch_size' (for us (5,11)) above the current pixel, symmetric to the left and right. A $5\times 11$ patch should e.g. go 5 pixels above, 5 to the left and 5 to the right of the current pixel.

```python
def extract_patch(img, row, col, patch_size):
    height, width = patch_size
    half_width = width // 2

    # Calculate the row range (above the pixel, including current pixel's row)
    start_row = row - height
    end_row = row  # +1 because slicing excludes the end index

    # Calculate the column range (symmetric around col)
    start_col = col - half_width
    end_col = col + half_width + 1  # +1 for slicing end index

    # Handle boundary conditions (pad if necessary)
    pad_top = max(0, -start_row)
    pad_left = max(0, -start_col)
    pad_bottom = max(0, end_row - img.shape[0])
    pad_right = max(0, end_col - img.shape[1])

    # Adjust start/end indices to be valid for slicing
    start_row_clamped = max(0, start_row)
    end_row_clamped = min(img.shape[0], end_row)
    start_col_clamped = max(0, start_col)
    end_col_clamped = min(img.shape[1], end_col)

    # Extract the patch (may be smaller than patch_size near edges)
    patch = img[start_row_clamped:end_row_clamped, start_col_clamped:end_col_clamped]

    # Pad the patch if near edges to keep the patch size consistent
    if any([pad_top, pad_bottom, pad_left, pad_right]):
        patch = np.pad(
            patch,
            ((pad_top, pad_bottom), (pad_left, pad_right)),
            mode='constant',
            constant_values=0
        )
    return patch
```

#### 2. Extract training patches for corrupted pixels
Theoretically, we could extract a patch for each individual pixel of all images (that have a sufficiently big neighbourhood).
To reduce the use of data and speed up our computations, we will just use the first 10 images of the training set.
We will further only extract patches for pixels that are currently corrupted in these training images. Since the corruption pattern is random, this should cover a sufficient number of different pixels.

There are different ways to extract the patches, the comments below highlight one way

```python
# For all images, go through each individual image (rows and cols) and find all corrupted pixels (value == 0)
# Then, extract the 5x11 patch directly above the corrupted pixel from the same image of the uncorrupted training set, 
# and add it to the new patch-based training set X;
# For the following prediction, also store the 'true' value of the pixel we want to reconstruct in the training labels set Y 
patch_size = (5, 11)

# Initialize lists to store patches (X) and target pixel values (Y)
patches_X = []
labels_Y = []

# Use first 10 training images
num_images = 10

for img_idx in range(num_images):
    corrupted_img = train_imgs_crpt[img_idx]  # corrupted image
    clean_img = train_imgs_clean[img_idx]    # corresponding clean image

    rows, cols = corrupted_img.shape
    
    for r in range(rows):
        for c in range(cols):
            # Check if pixel is corrupted (assuming corrupted pixels have value 0)
            if corrupted_img[r, c] == 0:
                # Extract patch directly above pixel (r, c) from clean image
                patch = extract_patch(clean_img, r, c, patch_size)
                
                # Flatten patch into 1D vector (optional, depends on model input)
                patch_flat = patch.flatten()
                
                # Add all the samples to an array (similar to the original training data)
                # Append to dataset
                patches_X.append(patch_flat)
                
                # Append true pixel value from clean image at (r,c)
                labels_Y.append(clean_img[r, c])

# Convert lists to numpy arrays
patches_X = np.array(patches_X)
labels_Y = np.array(labels_Y)

print(f"Extracted {patches_X.shape[0]} patches of size {patch_size} from first {num_images} images.")
print(f"X shape: {patches_X.shape}, Y shape: {labels_Y.shape}")
```

#### 3. Fitting the regression model to the information of the image patches

```python
# Retrieve coefficients using our linear regression fit function from Task 3
# Retrieve the regression coefficients to reconstruct images from corrupted ones
beta_patch = regression_fit(patches_X, labels_Y)
```

#### 4. Prediction / Reconstruction for ONE image of your choice
Prediction function is slightly different, since we cannot just directly predict ALL y-values! To predict a y-value, we need to have access to the window of size 5x11 right above the pixel we want to reconstruct. Intuitively, this can only be done for the first row of corrupted pixels, but not further (since the windows for the 2nd row would include corrupted pixels of the first row)! --> To solve this problem, we will reconstruct the image pixel-by-pixel using the reconstructed results of row1 to reconstruct row 2, and so on.


```python
import cv2 # Just for guassian blur in task 5
# Let's predict for one specific test image! 
img_idx_test = 8
corrupted_img_test = test_imgs_crpt_rdm[img_idx_test]
original_img_test = test_imgs_clean[img_idx_test]

# Create a copy of the corrupted image to reconstruct
reconstructed_img = corrupted_img_test.copy()

rows, cols = corrupted_img_test.shape
patch_height, patch_width = patch_size

for r in range(rows):
    for c in range(cols):
        # Skip non-corrupted pixels
        if corrupted_img_test[r, c] != 0:
            continue
        
        # We need (r - patch_height + 1) >= 0 for the patch to be valid
        if r - patch_height + 1 < 0:
            continue  # Cannot extract valid patch above this pixel
        
        # Extract patch from current reconstruction so far
        patch = extract_patch(reconstructed_img, r, c, patch_size)
        # patch = clean_patch(patch)
        patch_flat = patch.flatten()

        # Add bias term for regression prediction
        patch_aug = np.hstack(([1], patch_flat))  # shape: (56,)
        
        # Predict pixel value using trained regression coefficients
        pred_value = patch_aug @ beta_patch  # should now return float
        pred_value = float(pred_value.item())

        # Clip value to valid image range
        pred_value = np.clip(pred_value, 0, 255)
        
        # Assign predicted value to reconstructed image
        reconstructed_img[r, c] = pred_value

psnr_score = PSNR(reconstructed_img, original_img_test)
print(f"PSNR of reconstructed image: {psnr_score:.2f} dB")

# Smooth the final reconstructed image
smoothed_img = cv2.GaussianBlur(reconstructed_img, ksize=(3, 3), sigmaX=1)

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.title("Original")
plt.imshow(original_img_test, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Corrupted")
plt.imshow(corrupted_img_test, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("Reconstructed")
plt.imshow(reconstructed_img, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Reconstructed + filtered")
plt.imshow(smoothed_img, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

#### Compare the results you achieve with this local approach. What do you observe when you look at the reconstruction quality in the top half vs. the one in the lower half? Why is this the case?
<br>
Comparing the reconstruction results of the local patch-based regression, there is a clear difference in quality between the top and bottom halves of the image.

The top half of the reconstructed image appears much cleaner and closer to the original, whereas the bottom half shows noticeably more noise and degraded reconstruction quality.

This occurs because the prediction for each corrupted pixel depends on the pixel values above it, using a patch that includes the previous rows. At the top of the image, most of these patch pixels are from the original uncorrupted image, so predictions are based on accurate data. As we move downward, the algorithm increasingly relies on previously reconstructed (predicted) pixels. These predictions are not perfect, so errors start to accumulate. This creates a compounding effect, where each new prediction is based partly on already imperfect estimates, resulting in a progressive degradation of quality further down the image.

Overall, the local method works better than the regression model. It only uses information from above to predict the current pixel, meaning early errors propagate. This fundamentally limits this reconstruction method when processing in a top-to-bottom fashion.
